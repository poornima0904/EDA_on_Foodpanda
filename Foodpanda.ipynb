# EDA On FoodPanda

```python
# import liabraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
#read csv file
data = pd.read_csv('foodpanda.csv')
data.head()
```

```python
data.columns
```

```python
data.shape
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
#drop duplicates
data.drop_duplicates(inplace=True)
data.shape
```

```python
#convert qunatity and price into numerical values
data['qunatity'] = pd.to_numeric(data['qunatity'], errors='coerce')
data['price'] = pd.to_numeric(data['price'], errors='coerce')
data.head()
```

```python
#replace object features null with NONE
data['customer_id'] = data['customer_id'].replace(np.nan, 'NONE')
data['gender'] = data['gender'].replace(np.nan, 'NONE')
data['age'] = data['age'].replace(np.nan, 'NONE')
data['city'] = data['city'].replace(np.nan, 'NONE')
data['Order_id'] = data['Order_id'].replace(np.nan, 'NONE')
data['restaurant_name'] = data['restaurant_name'].replace(np.nan, 'NONE')
data['dish_name'] = data['dish_name'].replace(np.nan, 'NONE')
data['category'] = data['category'].replace(np.nan, 'NONE')
data['qunatity'] = data['qunatity'].replace(np.nan, 'NONE')
data['price'] = data['price'].replace(np.nan, 'NONE')
data['payment_method'] = data['payment_method'].replace(np.nan, 'NONE')
data['order_frequency'] = data['order_frequency'].replace(np.nan, 'NONE')
data['loyality_points'] = data['loyality_points'].replace(np.nan, 'NONE')
data['churned'] = data['churned'].replace(np.nan, 'NONE')
data['rating'] = data['rating'].replace(np.nan, 'NONE')
data['delivery_status'] = data['delivery_status'].replace(np.nan, 'NONE')
data.isnull().sum()
```

```python
data['price'] = data['price'].replace(np.nan, 0)
data.head()
```

# Data Visualization on Foodpanda

```python
sns.scatterplot(x=data['dish_name'], y=data['rating'])
plt.show()
```

```python
plt.figure(figsize=(20,6))
sns.countplot(x="price",data=data,palette="Set2")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(16,5))
sns.countplot(x='restaurant_name',data=data,palette="Set2")
plt.xticks(rotation=90)
plt.show()
```

```python
# dist plot on order_frequency
plt.figure(figsize=(20,9))
sns.distplot(data['order_frequency'],color='b')
plt.show()
```

```python
# dist plot on price
plt.figure(figsize=(20,9))
sns.distplot(data['price'],color='y')
plt.show()
```

```python
# line plot
plt.figure(figsize=(16,6))
plt.plot(data['rating'], data['payment_method'], color='r')
plt.show()
```
