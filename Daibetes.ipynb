```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('diabetes.csv')
data.head()
```

```python
data.describe()
```

```python
#drop duplicates
data.drop_duplicates(inplace=True)
data.shape
```

```python
# convert Pregnancies and  Insulin to numerical type
data['Pregnancies'] = pd.to_numeric(data['Pregnancies'], errors='coerce')
data['Insulin'] = pd.to_numeric(data['Insulin'], errors='coerce')
data.head()
```

```python
# replace Pregnancies and Insulin null values with 0
data['Pregnancies'] = data['Pregnancies'].replace(np.nan, 0)
data['Insulin'] = data['Insulin'].replace(np.nan, 0)
data.isna().sum()
```

```python
# replace object features null with NONE
data['Pregnancies'] = data['Pregnancies'].replace(np.nan, 'NONE')
data['Glucose'] = data['Glucose'].replace(np.nan, 'NONE')
data['BloodPressure'] = data['BloodPressure'].replace(np.nan, 'NONE')
data['SkinThickness'] = data['SkinThickness'].replace(np.nan, 'NONE')
data['Insulin'] = data['Insulin'].replace(np.nan, 'NONE')
data['DiabetesPedigreeFunction'] = data['DiabetesPedigreeFunction'].replace(np.nan, 'NONE')
data['Age'] = data['Age'].replace(np.nan, 'NONE')
data['Outcome'] = data['Outcome'].replace(np.nan, 'NONE')
data.isnull().sum()
```

# Data Visualization

```python
plt.figure(figsize=(20,6))
sns.countplot("Pregnancies",data=data,palette="Set2")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(16,5))
sns.countplot(x='BloodPressure',data=data,palette="Set2")
plt.xticks(rotation=90)
plt.show()
```

```python
# dist plot on Insulin
plt.figure(figsize=(16,5))
sns.distplot(data['Insulin'],color='g',bins=50)
plt.show()
```

```python
# dist plot on Age
plt.figure(figsize=(16,5))
sns.distplot(data['Age'],color='b')
plt.show()
```

```python
# line plot
plt.figure(figsize=(16,6))
plt.plot(data['DiabetesPedigreeFunction'], data['Insulin'], color='r')
plt.show()
```

```python
plt.figure(figsize=(16,5))
sns.scatterplot(x=data.Insulin, y=data.BMI)
plt.show()
```

```python
plt.figure(figsize=(16,5))
sns.scatterplot(x=data.BMI, y=data.Outcome, color="indigo")
plt.show()
```
